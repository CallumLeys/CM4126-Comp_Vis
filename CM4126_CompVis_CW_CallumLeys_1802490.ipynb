## 2.1 Research

1 Description

The paper ‘Deep virtual adversarial self-training with consistency regularisation for semi-supervised medical image classification’ (Wang et al., 2021) Introduces the problem of having limited annotated medical data used for deep neural network training, requiring extensive time and resources to generate this data.
The paper introduces a semi-supervised classification method for medical images that utilises deep virtual adversarial self-training (VAST) with consistency regularisation (CR) to combat the problem.
VAST is a technique for semi-supervised learning that combines both virtual adversarial training (VAT) which involves identifying small changes/modifications in the input data to help improve the output as the model would become more robust on small perturbations, with self-training, which involves using model’s predictions to produce new labelled data, this data is then re-trained on the model iteratively where it is believed to improve performance.
VAST also combines with CR which describes a method of implementing the model’s predictions on perturbed versions of the input data, therefore improving consistency on the model’s original data. Overall, this helps to ensure the model is robust and is not influenced due to variations in the input data.


2 Evaluation

2.1 Dataset

To provide a more justified analysis, the paper’s approach is performed on a private breast ultrasound dataset (BRUS) and a publicly accessible optical coherent tomography (OCT) dataset to allow for evaluation on datasets with different modalities (binary classification vs multi-class). The data is taken from real-world medical data, the Sun Yat-sen University Cancer Center for BRUS and multiple institutes for OCT (Kermany et al., 2018), so because it uses a diverse range of institutes, it helps replicate the various real-life examples that the model may work on.
BRUS has a total of 39,904 images of 8,079 lesions from 7,362 patients. 22,026 images are labelled where 14,557 are malignant and 7,469 are benign, the 17,878 rest are unlabelled.
OCT has 4 classes, CNV (37,455 images), DME (11,598 images), DRUSEN (8,866 images) which are diseases, and a NORMAL type (51,390 images). There are considerably more diverse images from OCT than BRUS which helps training and replication of the real world, however, due to training time constraints, we will be using smaller samples of the OCT dataset for the rest of the coursework.
The model determines the hyper-parameters by utilising a grid search.

2.2 Methods Presented

The approach of combining VAST and CN is a novel approach to semi-supervised medical image classification. The paper explains that the study (Sohn et al., 2020) showed that by simply combining semi-supervised approaches, benchmark results can be achieved. This was used as inspiration for the paper to combine different methods (VAST and CR) in the context of medical image classification, overall, this helps prove the proposed method is a novel, robust and simple approach to semi-supervised deep learning.
Due to a lack of extensive work on VAT in literature as well as no previous work that combines these methods, the technique is underdeveloped, yet interesting, leading to novel results.

2.3 Results

The paper compares evaluations from other related state-of-the-art methods.
The results show that the proposed approach achieves state-of-the-art performance compared to other semi-supervised learning methods and can improve classification accuracy when only a limited amount of labelled data is available.
The model provided by Huang et al., 2018 was used as a baseline, whereas multiple other related state-of-the-art models such as FixMatch from Sohn et al., 2020 were used in the comparison, with metrics such as accuracy, macro recall/precision/F1 and kappa scores being used.
It was found that the proposed model predicting the OTC dataset outperformed the models on every metric as seen in table 1. This proves the impressive nature of the method and can be used in further work in this domain.

3 Limitations

Although the method proposed is successful, the paper addresses its limitations. One is that the established method assumes that the labelled/unlabelled data are from the same distribution, however, this does not translate to real-life practice as it may come from differing institutions, which may have different software, hardware, and protocols. This results in a domain shift problem. The paper addresses this issue by proposing that the use of advanced domain adaption learning such as generative adversarial networks (GANs) which has shown promising results in the medical domain (Chen, Dou, Chen, Heng 2018, Wang, Yu, Li, Yang, Fu, Heng, 2019) and suggests that for future work, to implement this work to combat this problem.

4 Conclusion

To conclude, the paper proposes a method that combines powerful semi-supervised learning techniques to help improve the accuracy of medical image classification. The method is used on benchmark datasets and is demonstrated as effective when it is shown to outperform several state-of-the-art classification methods. While the results are promising, acknowledgement is made of domain shift issues and the paper suggests domain adaption techniques, such as GANs, as future work.

```python
# using the directory, we can collect the images and store into an np array
def get_image_data(data_dir):
    data = [] 
    for label in labels: 
        path = os.path.join(data_dir, label)
        class_num = labels.index(label)
        for img in os.listdir(path):
            try:
                img_arr = cv2.imread(os.path.join(path, img))
                resized_arr = cv2.resize(img_arr, (img_size, img_size)) # reshaping images to preferred size
                data.append([resized_arr, class_num])
            except Exception as e:
                print(e)
    return np.array(data)
```

```python
def equlize_hist_ (image):
  _image = np.asarray(image)
  image_eq = exposure.equalize_hist(_image)
  return image_eq
```

```python
def expose_imgae(path, label):
  # Choose normal random image: generate random number between 1 and  the number of normal images in the training set
  random_img_ind= np.random.randint(0,len(os.listdir(path))) 
  # image file name 
  img_expose_name = os.listdir(path)[random_img_ind]

  # path to the image
  img_expose_address = path+img_expose_name

  # load image 
  img_expose = Image.open(img_expose_address)
  img = np.asarray(img_expose)
  image_eq = equlize_hist_(img)

  figure1 = plt.figure(figsize= (16,16))
  img_1 = figure1.add_subplot(1,2,1)
  img_plot = plt.imshow(img, cmap = 'gray')
  img_1.set_title(label)
  plt.axis("off")

  img2 = figure1.add_subplot(1, 2, 2)
  img_plot = plt.imshow(image_eq, cmap = 'gray')
  img2.set_title( 'after HE')
  plt.axis("off") 
```

```python
# Model setup
model = Sequential()
model.add(Conv2D(32, (3,3), strides = 1, padding = 'same', activation = 'relu', input_shape = (img_size, img_size, 3)))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2), strides = 2, padding = 'same'))

model.add(Conv2D(64, (3,3), strides = 1, padding = 'same', activation = 'relu', input_shape = (img_size, img_size, 3)))
model.add(Dropout(0.1))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2), strides = 2, padding = 'same'))

model.add(Conv2D(64, (3,3), strides = 1, padding = 'same', activation = 'relu', input_shape = (img_size, img_size, 3)))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2), strides = 2, padding = 'same'))

model.add(Conv2D(128, (3,3), strides = 1, padding = 'same', activation = 'relu', input_shape = (img_size, img_size, 3)))
model.add(Dropout(0.2))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2), strides = 2, padding = 'same'))

model.add(Conv2D(256, (3,3), strides = 1, padding = 'same', activation = 'relu', input_shape = (img_size, img_size, 3)))
model.add(Dropout(0.2))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2), strides = 2, padding = 'same'))
model.add(Flatten())
model.add(Dense(units = 128, activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(units = 4, activation = 'softmax')) # 4 classes and softmax (commonly used in multi-class models)

model.compile(
              optimizer = 'rmsprop', loss = 'categorical_crossentropy', # categorical_crossentropy for multi class
              metrics = ['accuracy']
              )
```

```python
# model summary (see the number of trainable parameters)
model.summary()
```

```python
# introduce the learning rate reduction on plateau, which will decrease the rate of learning if the validation accuracy is on a plateau
learning_rate_reduction = ReduceLROnPlateau(monitor = 'val_accuracy', patience = 2, verbose = 1, factor = 0.3, min_lr = 0.000001)
```

We have now sorted the data and built the model accordingly, we will make it flow through several layers such as convid layers with relu activation functions and pooling layers until we reach the output layer which will consist of 4 nodes and will follow through a softmax activation function due to the multi-class output as it can give probability scores for each class [(Google Machine Learning - Multi-Class Neural Networks: Softmax)](https://developers.google.com/machine-learning/crash-course/multi-class-neural-networks/softmax#:~:text=Softmax%20extends%20this%20idea%20into,quickly%20than%20it%20otherwise%20would).

ReLu was chosen becasue it is efficient at combatting the vanishing gradient problem as the gradients of other functions such as sigmoid tends to become increasingly small after time. [(lin et al,. 2018)](https://www.sciencedirect.com/science/article/pii/S1877050918306197).

The reduce learning rate on plateau function is used to improve the learning of the model as the learning can stagnate. Reducing the learning rate is a key factor for the model as it can help monitor the performance and try to modify the learning rate to improve the model. [(johny et al., 2021)](https://www.hindawi.com/journals/cmmm/2021/5557168/) found that having a dynamic learning rate caused their model to reach 91.84% validation accuracy in less ephochs compared to static learning rate models, suggesting that we should use this method too.

Finally, batch normalization was used to help for more independant learning for each layer, what this means is for each layer, it will normalise the output before following through the next layer. this helps for avoiding overfitting the model [(liu et al., 2022)](https://www.nature.com/articles/s41598-022-18402-6#:~:text=CNN%20combines%20inception%2Dresidual%20modules,over%2Dfitting%20of%20the%20model) showed that using batchnormalisation along with other techniques of a 4-class model, 99.53% accuracy could be achieved.

## 4.3 Re-purposed Baseline CNN

With the results from the Transfer-Learning turning out dissapointing compared to the baseline, we move to building a 'traditional' CNN similar to the baseline and plan to augment it to provide better metrics.

```python
# Model setup
from keras.layers import LeakyReLU

model = Sequential()
model.add(Conv2D(32, (3,3), strides = 1, padding = 'same', activation=keras.layers.LeakyReLU(alpha=0.3), input_shape = (img_size, img_size, 3)))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2), strides = 2, padding = 'same'))

model.add(Conv2D(64, (3,3), strides = 1, padding = 'same', activation=keras.layers.LeakyReLU(alpha=0.3), input_shape = (img_size, img_size, 3)))
model.add(Dropout(0.1))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2), strides = 2, padding = 'same'))

model.add(Conv2D(64, (3,3), strides = 1, padding = 'same', activation=keras.layers.LeakyReLU(alpha=0.3), input_shape = (img_size, img_size, 3)))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2), strides = 2, padding = 'same'))

model.add(Conv2D(128, (3,3), strides = 1, padding = 'same', activation=keras.layers.LeakyReLU(alpha=0.3), input_shape = (img_size, img_size, 3)))
model.add(Dropout(0.2))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2), strides = 2, padding = 'same'))

model.add(Conv2D(256, (3,3), strides = 1, padding = 'same', activation=keras.layers.LeakyReLU(alpha=0.3), input_shape = (img_size, img_size, 3)))
model.add(Dropout(0.2))
model.add(BatchNormalization())
model.add(MaxPool2D((2,2), strides = 2, padding = 'same'))
model.add(Flatten())
model.add(Dense(units = 128, activation=keras.layers.LeakyReLU(alpha=0.3)))
model.add(Dropout(0.2))
model.add(Dense(units = 4, activation = 'softmax')) # 4 classes and softmax (commonly used in multi-class models)

model.compile(
              optimizer = 'adam', loss = 'categorical_crossentropy', # categorical_crossentropy for multi class
              metrics = ['accuracy']
              )
```

For this model, we introduce the adam optimiser and the utalize the Leaky ReLu funcion, justifications for these choices are explaine further on in the report.
